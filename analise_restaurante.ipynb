```python
import pandas as pd

dados = {
    "Data": ["2024-01-01","2024-01-01","2024-01-01","2024-01-02","2024-01-02"],
    "Categoria": ["Vendas Balcão","Delivery","Fornecedor","Bebidas","Salários"],
    "Tipo": ["Receita","Receita","Despesa","Receita","Despesa"],
    "Valor": [1200,800,500,600,2000],
    "Forma_Pagamento": ["Cartão","Pix","Pix","Dinheiro","Transferência"]
}

df = pd.DataFrame(dados)
df
```

```python
receitas = df[df['Tipo'] == 'Receita']
despesas = df[df['Tipo'] == 'Despesa']

total_receitas = receitas['Valor'].sum()
total_despesas = despesas['Valor'].sum()
lucro = total_receitas - total_despesas

print("Receitas:", total_receitas)
print("Despesas:", total_despesas)
print("Lucro:", lucro)
```

```python
receita_categoria = receitas.groupby('Categoria')['Valor'].sum()
despesa_categoria = despesas.groupby('Categoria')['Valor'].sum()

print(receita_categoria)
print(despesa_categoria)
```

```python
import matplotlib.pyplot as plt

receita_categoria.plot(kind='bar', title='Receita por Categoria')
plt.show()
```
